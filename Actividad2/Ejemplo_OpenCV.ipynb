# Mi primer proyecto de vision computacional
Este primer proyecto es un ehemplo de como utilizar OpenCV de manera efectiva

```python
# Importar librerias
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
bgr = cv2.imread('foto.webp')
print("El tama√±o de la imagen es: ", bgr.shape)

# Convertir a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

# Mostrar imagen
plt.imshow(rgb)
```

```python
# Mostrar la imagen a escala de grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
# Mostrar las componentes de color de la imagen
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(bgr[:,:,0], cmap='gray')
plt.title('Canal Azul')
plt.subplot(1,3,2) 
plt.imshow(bgr[:,:,1], cmap='gray')
plt.title('Canal Verde')
plt.subplot(1,3,3) 
plt.imshow(bgr[:,:,2], cmap='gray')
plt.title('Canal Rojo')
```

```python
# Mostrar las componentes HLS de la imagen
hls = cv2.cvtColor(bgr, cv2.COLOR_BGR2HLS)
plt.figure(figsize=(12,4))
plt.subplot(1,3,1)
plt.imshow(hls[:,:,0], cmap='gray')
plt.title('Canal H')
plt.subplot(1,3,2)
plt.imshow(hls[:,:,1], cmap='gray')
plt.title('Canal L')
plt.subplot(1,3,3)
plt.imshow(hls[:,:,2], cmap='gray')
plt.title('Canal S')
plt.show()
```
